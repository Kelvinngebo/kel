```python
# Combine all columns into a single column by stacking them vertically
all_centres_column = pd.concat([df[0], df[1], df[2]], ignore_index=True)
print(all_centres_column)
```

```python
# Optionally, turn it into a DataFrame
all_exam_centres_df = pd.DataFrame(all_centres_column, columns=['all_exam_centres'])
all_exam_centres_df
```

```python
# Step 1: Remove NaN or missing values
all_exam_centres_df = all_exam_centres_df.dropna()
all_exam_centres_df
```

```python
# Filter rows where the 'all_exam_centres' value starts with 'S'
school_centres_with_s = all_exam_centres_df[all_exam_centres_df['all_exam_centres'].str.startswith('S')].reset_index(drop=True)

school_centres_with_s.head(15)
```

```python
len(school_centres_with_s)
```

```python
# Display the conclusion using HTML
display(HTML(f"""
    <h3 style="color:green;">✅ Conclusion:</h3>
    <p>A total of <strong>{len(school_centres_with_s)} schools</strong> were successfully registered for the
    <strong>CSEE 2024 Examination</strong>.</p>
"""))
```

```python
# Filter rows that mention both 'Private' and 'Dar es Salaam'
dar_private_resit_centres = all_exam_centres_df[
    all_exam_centres_df['all_exam_centres'].str.contains("Dar es Salaam", case=False) &
    all_exam_centres_df['all_exam_centres'].str.contains("Private", case=False)
].reset_index(drop=True)

# Display top 10 to verify
dar_private_resit_centres.head(10)

# Show the number of centers
print(f"Total Private Resitting Centers in Dar es Salaam: {len(dar_private_resit_centres)}")
```
